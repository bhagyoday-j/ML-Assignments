```python
import numpy as np
import matplotlib.pyplot as plt 

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints, color = "green", marker = "o")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("My first graph!")
plt.show()
```

```python
students = np.array(['A', 'B', 'C', 'D'])
marks = np.array([90, 80, 70, 60])

plt.bar(students, marks, color = "yellow")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks of students")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y1 = [10,20,25,30]
y2 = [5,15,20,25]

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title("First Plot")

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("Second Plot")

plt.show()
```
